### plot the kernel error

```python
plt.figure()
for nbit in [1, 2, 4, 8]:
    val = []
    for budget in [32, 64, 128, 256, 512, 1024]:
        key = "lp_rff_budget_" + str(budget) + "_nbit_" + str(nbit)
        val.append(results[key][0] )
    plt.semilogy(val, label="lp rff " + str(nbit) + " bit")
val = []
for budget in [32, 64, 128, 256, 512, 1024]:
    val_tmp = []
    for mu in [1.0, 2.0, 3.0, 5.0, 10.0, 0.2, 0.33, 0.5]:
        key = "lp_pca_rff_budget_" + str(budget) + "_mu_" + str(mu)
        val_tmp.append(results[key][0] )
    val.append(min(val_tmp))
plt.semilogy(val, label="lp pca rff ")
plt.grid()
plt.legend()
plt.title("kernel approx error to exact kernel")
plt.show()
```
